```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
dataset = pd.read_csv('iris.csv')
dataset
```

```python
cols = dataset.columns[:-1]
```

```python
for col in cols:
    sns.histplot(data=dataset, kde=True, x=col, linewidth=2)
    plt.figure(figsize=(10, 6))
    plt.show()
```

```python
for col in cols:
    sns.boxplot(data=dataset, x=col)
    plt.show() 
```
